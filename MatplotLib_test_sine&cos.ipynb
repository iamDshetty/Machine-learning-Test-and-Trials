```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,2*np.pi,100)
sin_x = np.sin(x)
cos_x = np.cos(x)
fig,ax = plt.subplots(figsize=(15,5))
_=ax.set_title("Sine wave")
#_=ax.plot(x,sin_x,marker="*",drawstyle="steps", markersize=2,markerfacecolor='white',markeredgecolor="green",markeredgewidth=2,color="skyblue",linewidth=2,linestyle="-.")
_=ax.plot(x,sin_x,drawstyle="steps", label="Sine",linestyle='-.')
_=ax.plot(x,cos_x,drawstyle="steps", label="Cos",linestyle='-.')
_=ax.plot(x,x, label="x",linestyle='-.')
_=ax.set_aspect('equal')
_=ax.legend(loc=0)

```
