```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#numerical data plots

#jointplot # 2 + more
#pairplot # whole df comparison
#distplot # single 

#Categorical data plots

#BOXPLOT - both categorical and numerical
#violin plot - combination of boxplot and kde
#countlot  - single
#barplot  - categorical vs numeric
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
df.corr() # -1 to +1 only numbers not categorical
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y="total_bill",data=df,kind='hex')  # kind = "hex" hexagonal shap  "reg" regression line
```

```python
sns.pairplot(df) #integer value
```

```python
sns.pairplot(df,hue='sex') #with categorical as hue  
```

```python
df['smoker'].value_counts()
```

```python
sns.distplot(df['tip'],kde=False,bins=10) #kernal density estimation = kde ==true it shows percentage
```

```python
sns.countplot('day',data=df)  
```

```python
sns.countplot(y='day',data=df)  
```

```python
sns.barplot(y='total_bill',x='day',data=df)
```

```python
sns.boxplot('smoker','total_bill',data=df)  #
```

```python
sns.boxplot('smoker','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,palette='rainbow',orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue="smoker",data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
