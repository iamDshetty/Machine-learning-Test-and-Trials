```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/story645/mpl_tutorial/master/notebooks/data/titanicfull.csv").dropna()
survived=df['survived'].value_counts()


fig,ax = plt.subplots()
_=ax.pie(x=survived.values,labels=['survived','died'],colors=["green","red"])
survived
```
