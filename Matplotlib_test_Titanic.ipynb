```python
 fig, ax = plt.subplots()
_=ax.plot(np.sort(df['age']),marker='o',markersize=1)
_=ax.set_ylabel("Ages")
_=ax.set_xlabel(r"Passengers $\theta$")
_=ax.set_yticks([15,25,55,64])
_=ax.set_yticklabels(['children','youth','adults','senior'])
```

```python
fig2,ax2 = plt.subplots()
_=ax2.set_title("Titanic")
_=ax2.plot(np.sort(df["fare"]),marker='o',linestyle="",markersize='10')
_=ax2.set_yscale("log")
_=ax2.set_ylabel("fares in Euros")
_=ax2.set_xlabel("passesngers")
_=ax2.set_yticks([3,12,30,870])
_=ax2.set_yticklabels(["Thirdclass","Second Class","First Class Berth","First Class Suite"])

```

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=2,constrained_layout=True)
_=fig.suptitle("NEXT To (1x2)")
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,constrained_layout=True,gridspec_kw={'width_ratios':[1,1]})
_=fig.suptitle("NEXT To (1x2)")

_=ax1.plot(np.sort(df["age"]),marker="o",markersize=5,linestyle="")
_=ax1.set_ylabel("Ages")
_=ax1.set_xlabel("passengers")
_=ax1.set_yticks([15,25,55,64])
_=ax1.set_yticklabels(["child","Young","Adult","Senior"])

_=ax2.plot(np.sort(df["fare"]),marker="o",markersize="5",linestyle="")
_=ax2.set_ylabel("Fares")
_=ax2.set_xlabel("Passengers")
_=ax2.set_yscale("log")
_=ax2.set_yticks([3,12,30,870])
_=ax2.set_yticklabels(["Third Class","Second Class","First Class Berth","First Class Suite"])



```
