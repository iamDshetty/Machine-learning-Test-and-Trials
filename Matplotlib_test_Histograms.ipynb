```python
#Matplotlib Histograms

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#matplotlib.__version__
#np.__version__
#pd.__version__

df = pd.read_csv("https://raw.githubusercontent.com/story645/mpl_tutorial/master/notebooks/data/titanicfull.csv").dropna()

fig,(ax) = plt.subplots()
#_=ax.hist("age",data=df)#,bins="auto")
#(counts,edges,_)=ax2.hist("age",data=df,bins="auto") 
(counts,edges,_) = ax.hist("age",data=df,bins=[0,15,25,55,64,199],edgecolor="black",color="white",hatch="\\")
_=ax.set_xticks([0,15,25,55,64,199])

```

```python
fig,ax=plt.subplots()
(density,edges,_) = ax.hist("age",data=df,color="white",edgecolor="black", hatch="//",bins="auto",density=True )
assert np.sum(np.diff(edges)*density) ==1
```

```python
fig,ax = plt.subplots()
(density,edges,_)=ax.hist("fare", data=df,color="lightblue",edgecolor="black",hatch="//",bins="auto")#bins=[0,3,12,30,870])
#_=ax.set_xticks([0,3,12,30,870])
#_=ax.set_xscale("log")
```

```python
#for loop histogram , group by gender
fig,(ax,ax2) = plt.subplots(ncols=2,constrained_layout=True)
for label,gdf in df.groupby('sex'):
    ax.hist("age",data=gdf,bins="auto",edgecolor="black" ,label=label,alpha=0.7)
    ax2.hist("age",data=gdf,bins="auto",edgecolor="black",label=label,alpha=0.7,density=True)
_=ax.legend()
_=ax.set(title='count',ylabel='N')
_=ax2.legend()
_=ax2.set(title='density',xlabel='age',ylabel='P')
```
