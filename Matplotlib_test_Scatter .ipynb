```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/story645/mpl_tutorial/master/notebooks/data/titanicfull.csv").dropna()
#df.head()

cmap = mcolors.ListedColormap(["blue","green"])

fig,ax = plt.subplots()
_=ax.scatter("age","fare",data=df,c="survived", cmap=cmap)      #(df['age'],df["fare"])  #color="lightgreen",
_=ax.set(xlabel="age",ylabel="fare",title='Titanic Passengers')



```

```python
fig,ax = plt.subplots()
cmap = mcolors.ListedColormap(["grey","purple"])
sc=ax.scatter("age","fare",data=df,c="survived", cmap=cmap,edgecolor='black',alpha=0.6)      #(df['age'],df["fare"])  #color="lightgreen",

_=ax.set(xlabel="age",ylabel="fare",title='Titanic Passengers')
#_=ax.legend(*sc.legend_elements())
_=ax.legend(handles=sc.legend_elements()[0],labels=['Died','Survived']) #returnes 2 value i.e.., handles & labels
```

```python
#bubble chart

fig,ax = plt.subplots()
cmap1 = mcolors.ListedColormap(["red","Green"])
SC=ax.scatter("age",'fare',data=df,edgecolor="black",alpha=0.5,cmap=cmap1,c="survived",s=df["pclass"]*25)
_=ax.set(xlabel="Age",ylabel="Fare",title="Titanic")
_=ax.legend(handles=SC.legend_elements()[0],labels=["Died","Survived"])

```
