```python
#Linear Regression Algorithm 

#Mathematical Calculation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)

data= pd.read_csv("D:/Chandrashekar S/New Volume/Machine learning/Scikit-learn/headbrain.csv")

x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values

mean_x = np.mean(x)
mean_y=np.mean(y)
n=len(x) #total number of value

fig,ax =plt.subplots()
_=ax.scatter(x,y)
_=ax.set(xlabel="Head Size(cm^3)",ylabel="Brain Weight(grams)")

data.head()
```

```python
#y=mx+C find Regression line

#find m & C

#m=Sumation of[(x-mean(x))(y-mean(y))]/sumation of[(x-mean(x))square]

numer = 0
denom=0
for i in range(n):
    numer +=(x[i]-mean_x)*(y[i]-mean_y)
    denom +=(x[i]-mean_x)**2
m=numer/denom
C=mean_y-(m*mean_x)
m,C
```

```python
max_x = np.max(x)+100
min_x = np.min(x)-100

X=np.linspace(min_x,max_x,1000)
Y=C+m*X

fig,ax =plt.subplots()
_=ax.scatter(x,y)
_=ax.plot(X,Y)
_=ax.set(xlabel="Head Size(cm^3)",ylabel="Brain Weight(grams)")
```

```python
ss_t = 0
ss_r = 0

for i in range(n):
    y_pred=m*x[i]+C
    ss_t +=(y[i]-mean_y)**2
    ss_r +=(y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
r2
    
```

```python
#linear Regression using Scikit Learn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x = x.reshape((n,1)) # reshape
reg = LinearRegression() # Creating model
reg = reg.fit(x,y) #fitting training data
y_predt = reg.predict(x) # predicting Y while is best fit line i..e linear regression line


#calculate R square value
mse = mean_squared_error(y,y_predt)
rmse = np.sqrt(mse)
r2_score = reg.score(x,y) #coeffients

rmse,r2_score




```
